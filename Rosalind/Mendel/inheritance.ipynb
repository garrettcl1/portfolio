```python
def mendel(k, m, n):
    #prob of recessive (complement)
    pop = k+m+n
    pairs = pop*(pop-1)
    nn = n * (n-1)
    nm = 2 * m * n
    mm = m * (m-1)
  
    
    p_mn = .5
    p_mm = .25
    p_nn = 1
    prob_recessive = (nn + .5*nm + .25*mm)/pairs
    prob = 1 - prob_recessive
    return prob
```
