```python
with open("rosalind_dna (1).txt", "r") as txt:
    contents = txt.read()

```

```python
print(contents)
```

```python
A = 0
C = 0
G = 0
T = 0
for x in contents:
    if x == "A":
        A += 1
    if x == "C":
        C += 1
    if x == "G":
        G += 1
    if x == "T":
        T += 1
print(A, C, G, T)
```
