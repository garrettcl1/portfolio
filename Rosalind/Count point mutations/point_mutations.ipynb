# point mutations problem

```python
with open("rosalind_hamm.txt", "r") as txt:
    fasta = txt.read()
    print(fasta)
```

```python
def hamm_d(text):
    sequences = text.strip().split('\n')
    
    seq1 = sequences[0]
    seq2 = sequences[1]
    
    return sum(c1 != c2 for c1, c2 in zip(seq1, seq2))
```

```python
hamm_d(fasta)
```
