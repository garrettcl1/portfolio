```python
import re

def gc(text):
    pattern = r'>([^\n>]+)\n?([ACGTU\n]+)'
    matches = re.findall(pattern, text)
    
    gc_content = {}
    for label, sequence in matches:
        label = label.strip()
        # Remove all whitespace/newlines from sequence
        sequence = sequence.replace('\n', '').replace('\r', '').strip()
        
        gc_count = sequence.count('G') + sequence.count('C')
        gc_percent = (gc_count / len(sequence)) * 100
        gc_content[label] = gc_percent
    
    max_label = max(gc_content, key=gc_content.get)
    return max_label, gc_content[max_label]

# Usage
with open("rosalind_gc.txt", "r") as txt:
    fasta = txt.read()

```

```python
gc(fasta)
```
