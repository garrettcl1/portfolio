```python
text = "AAAACCCGGT"
with open("output.txt", "w") as file:
    file.write(text)
```

```python
with open("output.txt", "r") as txt:
    dna = txt.read()
print(dna)
```

```python
def complement(dna):
    trans = str.maketrans('ATCG', 'TAGC')
    return dna.translate(trans)[::-1]
```

```python
r_comp = complement(dna)
print(r_comp)
print(r_comp == 'ACCGGGTTTT')
```
